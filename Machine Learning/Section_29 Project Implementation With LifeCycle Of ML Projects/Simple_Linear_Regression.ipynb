### Simple Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Read the data set
df=pd.read_csv('height_weight.csv')
df.head()
```

```python
plt.scatter(df['Weight'],df['Height'])
plt.xlabel('Weight')
plt.ylabel('Height')
```

```python
# divide  our dstaset into independent and dependent features
X=df[['Weight']] ##indedpendent features
y=df['Height'] ##dependent features
```

```python
## Train test split
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=.20,random_state=42)
```

```python
X.shape
```

```python
X_train.shape,X_test.shape ,y_train.shape,y_test.shape
```

```python
## standardizaing the dataset train independent data
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

```python
plt.scatter(X_train,y_train)
```

```python
## train the Simple Linear Regression model
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
print("the slope or coeff of weight is",regressor.coef_)
print("Intercept", regressor.intercept_)
```

```python
plt.scatter(X_train,y_train)
plt.plot(X_train,regressor.predict(X_train),'r')
```

### prediction of train data
1.predicted height output= intercept +coef_(Weights)

2.y_pred_train =157.5 + 17.03(X_train)

### prediction of test data
1.predicted height output= intercept +coef_(Weights)

2.y_pred_test =157.5 + 17.03(X_test)

```python
y_pred_test=regressor.predict(X_test_scaled)

```

```python
y_pred_test, y_test
```

```python
plt.scatter(X_test_scaled,y_test)
plt.plot(X_test_scaled,regressor.predict(X_test_scaled),'r')
```

### Performance Metrics

### MSE, MAE, RMSE

### R squeare and adjusted R square

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
mse=mean_squared_error(y_test,y_pred_test)
mae=mean_absolute_error(y_test,y_pred_test)
rmse=np.sqrt(mse)
print(mse)
print(mae)
print(rmse)
```

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred_test)
```
